# Preparing the Data
Separando os dados em features e label

```python
nota_raw = d['Nota do Consumidor']
features_raw = d.drop('Nota do Consumidor',axis=1)
```

Normalizando features numéricas com MinMaxScaler()

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
f_numericas = ['Mês Abertura', 'Mês Finalização', 'Mês Resposta', 'Tempo Resposta']

features_minmax = pd.DataFrame(data=features_raw)
features_minmax[f_numericas] = scaler.fit_transform(features_minmax[f_numericas])

display(features_minmax.head())
```

Transformando features não numéricas em numéricas com get_dummies()

```python
features_final = pd.get_dummies(features_minmax)
```

```python
features_final.columns
```

```python
print(features_final)
```

```python
len(features_final)
```

Transformando nota_raw em int

```python
nota_final = nota_raw.astype(int)
```

Cross Validation

```python
# Import train_test_split
from sklearn.cross_validation import train_test_split

# Split the 'features' and 'income' data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features_final, 
                                                    nota_final, 
                                                    test_size = 0.2, 
                                                    random_state = 101)

from sklearn.feature_selection import SelectPercentile
select = SelectPercentile(percentile=50)
select.fit(X_train,y_train)
X_train_selected = select.transform(X_train)
X_test_selected = select.transform(X_test)

# Show the results of the split
print("Training set has {} samples.".format(X_train.shape[0]))
print("Testing set has {} samples.".format(X_test.shape[0]))
```

```python
mask = select.scores_ #list of booleans
principais_features = [] # The list of your K best features

for bool, feature in zip(mask, features_final):
    if bool:
        principais_features.append(feature)

results = pd.DataFrame(select.scores_,principais_features)
results.columns = ['Importancia']
results.sort_values(['Importancia'],ascending=False)
```

# Criação de modelos

Criar um modelo Naive para utiliza-lo como benchmark, afim de comparar com os outros modelos indicados na proposta do projeto.

```python
from sklearn.naive_bayes import MultinomialNB

from sklearn.metrics import f1_score

model = MultinomialNB().fit(X_train_selected, y_train)  
predicted = model.predict(X_test_selected)
print (f1_score(y_test,predicted,average="micro"))
```

Criando um modelo GradientBoostingClassifier

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.multiclass import OneVsRestClassifier

clf_gbc = OneVsRestClassifier(GradientBoostingClassifier(random_state = 101))

#Treinando o modelo 
gbc_fit = clf_gbc.fit(X_train_selected, y_train)

pred = gbc_fit.predict(X_test_selected)

score = f1_score(y_test, pred, average="micro")
print("f1-score for test set: {:f}".format(score)) 
```

Criando um modelo LinearSVC

```python
from sklearn.svm import LinearSVC

clf_svc = OneVsRestClassifier(LinearSVC(random_state = 101))

#Treinando o modelo com o GridSearch Ajustado
svc_fit = clf_svc.fit(X_train_selected, y_train)

pred = svc_fit.predict(X_test_selected)

score = f1_score(y_test, pred, average="micro")
print("f1-score for test set: {:f}".format(score)) 
```

Criando um modelo GradientBoostingClassifier com GridSearch para parametrizar o modelo

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.grid_search import GridSearchCV
from sklearn.metrics import make_scorer
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import f1_score
from sklearn import metrics

#Criando o modelo Adaboost
#'estimator__learning_rate','estimator__n_estimators','estimator__min_samples_split','estimator__max_depth'
#'estimator__n_estimators': [50,60,70,80,90,100] = 60
#'estimator__learning_rate': [0.1, 0.5, 1] = 0.1
#'estimator__min_samples_split':[2,5,8] = 2
#'estimator__max_depth':[3,5,8] = 5
#'estimator__min_samples_leaf':[30,40,50,60,70] = 40
#'estimator__subsample':[0.2,0.5,0.8] = 0.8

clf_gbc = OneVsRestClassifier(GradientBoostingClassifier(learning_rate = 0.1, n_estimators = 60,min_samples_split = 2,min_samples_leaf = 40, max_depth = 5))

#Criando os parametros do GridSearch
param_test1 = {'estimator__subsample':[0.5,0.8]}

scor = make_scorer(f1_score,average="micro")

#Executar o GridSearch
gbc_gs = GridSearchCV(clf_gbc,param_grid = param_test1,scoring =scor)

#Treinando o modelo 
gbc_fit = gbc_gs.fit(X_train_selected, y_train)

best_clf = gbc_fit.best_estimator_

pred = best_clf.predict(X_test_selected)

print("f1-score for test set: {:f}".format(f1_score(y_test,pred,average="micro"))) 
```

```python
from sklearn.svm import LinearSVC

clf_svc = OneVsRestClassifier(LinearSVC(random_state = 101))

param_test1 = {'estimator__C': [0.1,0.5,1]}

scor = make_scorer(f1_score,average="micro")

clf_svc = GridSearchCV(clf_svc,param_grid = param_test1,scoring =scor)

modelo_fit = clf_svc.fit(X_train_selected, y_train)

best_clf = modelo_fit.best_estimator_

pred = best_clf.predict(X_test_selected)

score = f1_score(y_test, pred, average="micro")
print("f1-score for test set: {:f}".format(score)) 
```
